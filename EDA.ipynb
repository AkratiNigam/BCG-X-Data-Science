
# Exploratory Data Analysis (EDA)

This notebook performs a complete EDA on:

- `client_data.csv`
- `price_data.csv`

The analysis includes:
- Data loading
- Data types inspection
- Missing value analysis
- Distribution analysis
- Basic statistical summaries
- Visualizations with interpretations


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load datasets
client_df = pd.read_csv('client_data (1).csv')
price_df = pd.read_csv('price_data (1).csv')

print("Client Data Shape:", client_df.shape)
print("Price Data Shape:", price_df.shape)
```

## 1. Data Types Inspection

```

print("Client Data Types:")
print(client_df.dtypes)

print("\nPrice Data Types:")
print(price_df.dtypes)

# Interpretation:
# This shows which columns are numerical, categorical (object), or dates (currently object).
# Date columns will need conversion for time-based analysis.
```

```

date_cols_client = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']
for col in date_cols_client:
    client_df[col] = pd.to_datetime(client_df[col], errors='coerce')

price_df['price_date'] = pd.to_datetime(price_df['price_date'], errors='coerce')

print("Date conversion completed.")
```

## 2. Missing Value Analysis

```

print("Missing Values in Client Data:")
print(client_df.isnull().sum())

print("\nMissing Values in Price Data:")
print(price_df.isnull().sum())

# Interpretation:
# Columns with high missing values may require imputation or removal depending on business relevance.
```

## 6. Correlation Analysis (Client Data)

```

corr = client_df.select_dtypes(include=np.number).corr()

plt.figure()
plt.imshow(corr)
plt.title("Correlation Matrix (Numerical Features)")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.show()

# Interpretation:
# The correlation matrix helps identify strongly correlated variables.
# Highly correlated features may cause multicollinearity in modeling.
```
